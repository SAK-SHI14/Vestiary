```python
color_map = {
    'red': 'warm', 'maroon': 'warm', 'pink': 'warm', 'orange': 'warm',
    'blue': 'cool', 'navy': 'cool', 'sky': 'cool', 'teal': 'cool', 'purple': 'cool',
    'green': 'natural', 'olive': 'natural',
    'black': 'neutral', 'white': 'neutral', 'grey': 'neutral', 'gray': 'neutral',
    'yellow': 'bright', 'gold': 'bright', 'beige': 'bright', 'brown': 'natural'
}
```

```python
def map_color_group(color):
    if pd.isna(color):
        return 'unknown'
    for key in color_map:
        if key in color.lower():
            return color_map[key]
    return 'other'
```

```python
data['color_group'] = data['baseColour'].apply(map_color_group)
```

```python
occasion_map = {
    'casual': 'daily wear',
    'formal': 'office wear',
    'party': 'special occasions',
    'workout': 'activewear',
    'sports': 'activewear',
    'festive': 'special occasions'
}
```

```python
def map_occasion(occasion):
    if pd.isna(occasion):
        return 'unspecified'
    for key in occasion_map:
        if key in occasion.lower():
            return occasion_map[key]
    return occasion.lower()

data['occasion_type'] = data['usage'].apply(map_occasion)
```

```python
def map_weather(occasion):
    if pd.isna(occasion):
        return 'any'
    occasion = occasion.lower()
    if 'workout' in occasion or 'sports' in occasion:
        return 'hot'
    elif 'party' in occasion or 'formal' in occasion:
        return 'cool'
    elif 'casual' in occasion:
        return 'hot'
    return 'any'
```

```python
data['weather_compatibility'] = data['usage'].apply(map_weather)
```
